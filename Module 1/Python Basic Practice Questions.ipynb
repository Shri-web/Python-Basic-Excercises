#### QUESTION 1

Perform Caesar Cipher technique

As input, give a String and a constant integer value. Find the encrypted text message for each of the following Strings.

Rules for the encryption are, for each character in string shift the character to the right of a specific fixed value that you are given as the input.
 
#### E(x) = (x + n) mod(26)


Input Format
    
    First line of input consist of String that you want to encrypt.

    Second line of the input consist of a fixed integer value to shift.
Output Format
    
    Output consist of Encrypted string.
Constraints
    
    0 < n < ∞

Sample Input 1
    
    This is the secret message
    10
Sample Output 1
    
    Drscxscxdroxcombodxwocckqo

```python
def encrypt(text,val):
    result = ""

    for i in range(len(text)):
        char = text[i]

        if (char.isupper()):
            result += chr((ord(char) + val-65) % 26 + 65)

        else:
            result += chr((ord(char) + val - 97) % 26 + 97)

    return result

text = input()
val = int(input())

print(encrypt(text,val))
```

#### QUESTION 5

Create a pascal triangle by getting the number of rows as an input.
    
Pascal Triangle concept:
    
Each number is the sum of the two numbers directly above it 
    
                                        1
    
                                1                 1
    
                      1                 2                 1
    
             1              3                     3              1
    
        1          4                   6                  4            1
    
Input Format 
    
    First line of input is no of rows.
    
Output Format 
    
    The output should be in list for each row
    
Sample input 1
    
    3
    [[1], [1, 1], [1, 2, 1]]
    
Sample input 2
    
    5
    [[1], [1, 1], [1, 2, 1], [1, 3, 3, 1], [1, 4, 6, 4, 1]]

```python
def Pascal(rows):
    List = [0]*rows
    for i in range(rows):
        if i == 0:
            List[i] = [1]
        elif i== 1:
            List[i] = [1,1]
        else:
            List[i] = [0]*(i+1)
            List[i][0] = List[i][i] = 1
            for j in range(1,i):
                List[i][j] = List[i-1][j-1]+List[i-1][j]
    return List

n = int (input())
print (Pascal(n))
```

#### QUESTION 10

We all know that computer understands the characters in 0’s and 1’s (binary). So your task is to convert a decimal number from the input to a binary number.
    
Input format:
    
    First line of input consist of the decimal number 
    
Output format:
    
    The first line consist of the binary equivalent of the decimal number.
    
Sample input:
    
    85
    
Sample output:
    
    1010101

```python
def dec_to_binary(n):
	bits = []
	
	while n > 0:
		bits.append(n%2)
		n = n // 2
	bits.reverse()
	
	binary = ''
	for bit in bits:
		binary += str(bit)
	return binary


num = int(input())
binary = dec_to_binary(num)
print(binary)
```

The longest common subsequence (LCS) is defined as the longest subsequence that is common to all the given sequences, provided that the elements of the subsequence are not required to occupy consecutive positions within the original sequences.
<br>
Implement a code using recursion to find the length of the least common subsequence (lcs).

Input Format
    
    First line consist of the string
    Second line consist of another string
Output Format
    
    Returns the length of the lcd
Sample Input 1
    
    abcdef
    abcdfghij
Sample Output 1
    
    5

```python
def lcs(X, Y, m, n):
 
    if m == 0 or n == 0:
       return 0;
    elif X[m-1] == Y[n-1]:
       return 1 + lcs(X, Y, m-1, n-1);
    else:
       return max(lcs(X, Y, m, n-1), lcs(X, Y, m-1, n));
 
 
X = input()
Y = input()
print (lcs(X, Y, len(X), len(Y)))
```

#### QUESTION 16

Create a programme that identify the common substring among all the strings in an array. Use functions to do so.

Input Format
    
    Enter the list of values from user
Output Format
    
    Returns the common sequence.
Sample Input 1
    
    example ex exclamation expect
Sample Output 1
    
    ex

```python
def longest_common_prefix(strings):
  if not strings:
    return ''
  
  for i, chars in enumerate(zip(*strings)):
    if len(set(chars)) > 1:
      return strings[0][:i]
  else:
    return min(strings)

strings =   list(map(str,input().strip().split()))
print(longest_common_prefix(strings))

```
